```python
import numpy as np
import opuslib
import IPython
channels = 1
rate = 16000
```

## opus encoding and decoding

```python
with open("../../SA2.wav", "rb") as file:
    fh=file.read(40000)
```

```python
IPython.display.Audio(np.fromstring(fh[1024:3*960*6+1024],np.int16),rate=16000)
```

```python
encoder=opuslib.Encoder(rate,channels,'audio')
```

```python
# can be 960, 960*3 etc. keep same in decoding step
encdata=[encoder.encode(fh[i*960*6+1024:(i+1)*960*6+1024],960*2) for i in range(3)]
```

```python
encdata
```

```python
[len(i) for i in encdata]
```

```python
decoder=opuslib.Decoder(rate,channels)
```

```python
decdata=[decoder.decode(i,960*2) for i in encdata]
```

```python
[len(i) for i in decdata]
```

> comment:   
>     1.almost decode to size of original audio track   
>     2.compression rate ~ 3840/200=19.2  

```python
IPython.display.Audio(np.fromstring(b''.join(decdata),np.int16),rate=16000)
```
